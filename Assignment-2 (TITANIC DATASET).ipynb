# Handle the Null Values

```python
pack["age"].fillna(pack["age"].median(), inplace=True)
```

```python
pack["embarked"].fillna(pack["embarked"].mode()[0], inplace=True)
```

```python
pack["deck"].fillna(pack["deck"].mode()[0], inplace=True)
```

```python
pack["embark_town"].fillna(pack["embark_town"].mode()[0], inplace=True)
```
